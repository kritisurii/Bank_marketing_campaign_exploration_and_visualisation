<h2> Data Visualization </h2>

```python
df.hist(bins=50, figsize=(20,15), color='r')
plt.show()
```

```python
#Blue-collar,Management and Technicians constitutes 60% percent of the population .
sns.set(style="darkgrid")
plt.figure(figsize=(15,12))
sns.countplot(x='job',hue='education',data=df)
plt.show()
```

```python
#Conclusion- single=min age and higher qualification, divorce = max age and less qualification
df4=df.pivot_table(index='marital',columns='education',values='age')
sns.heatmap(df4)
```

```python
#Conclusion: People having balance more than 2500 have subscribed to term deposits in month of December 
df11=df.pivot_table(index='y',columns='month',values='balance')
sns.heatmap(df11,linewidth=1,cmap="Blues")
```

```python
#To check the relationship between job and balance
sns.stripplot(x='job',y='balance',data=df)
```

```python
#Conclusion: It can be inferred that people with management jobs have comparatively higher balance as compared to students and unknown jobs.
plt.figure(figsize=(12,10))
sns.stripplot(x='job',y='balance',data=df)
```

```python
plt.figure(figsize=(12,10))
sns.stripplot(x='job',y='age',data=df,hue='loan',jitter=True)
#Conclusion:It can be inferred that little or nil amount of students have taken loan and other jobs have equal amounts of loantakers.
```

```python
#To analyse Job Distribution with Deposit status 
plt.figure(figsize=(18,10))
sns.violinplot(x="balance",y="job",hue='y',data=df)
plt.title("Job distribution with Deposit status",fontsize=16)
plt.show()
```

```python
df10=df.pivot_table(index='job',columns='marital',values='age')
sns.heatmap(df10,linewidth=1)
#To understand how job and age can affect marital status?
```
